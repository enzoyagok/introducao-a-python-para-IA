# ***5 . exercício - Explique modelo.fit***(X, Y)


* O comando model.fit(X, y) treina o modelo de Machine Learning utilizando os dados de entrada (X) e os resultados esperados (y). Durante esse processo, o modelo aprende padrões nos dados para poder realizar previsões em novos casos.

# ***Desafio***

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier

dados = {
    "horas_estudo": [1, 2, 3, 4, 5, 6],
    "aprovado": [0, 0, 0, 1, 1, 1]
}

# Cria a tabela
df = pd.DataFrame(dados)

# Entradas
X = df[["horas_estudo"]]

# Saídas
y = df["aprovado"]

# Cria o modelo
modelo = DecisionTreeClassifier()

# Treina o modelo
modelo.fit(X, y)

# Novo aluno (5 horas de estudo)
novo_aluno = [[5]]

# Faz previsão
resultado = modelo.predict(novo_aluno)

# Exibe o resultado
if resultado[0] == 1:
    print("O aluno será aprovado.")
else:
    print("O aluno será reprovado.")
```
